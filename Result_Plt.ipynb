```python
import ast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

# E1 visulization here

```python
experiment_result_1 = pd.read_csv('/content/drive/MyDrive/experiment_result_1.csv')
```

```python
# experiment_result_1.head()
```

```python
experiment_result_1['train_loss'] = experiment_result_1['train_loss'].apply(ast.literal_eval)
experiment_result_1['valid_loss'] = experiment_result_1['valid_loss'].apply(ast.literal_eval)
experiment_result_1['train_acc'] = experiment_result_1['train_acc'].apply(ast.literal_eval)
experiment_result_1['valid_acc'] = experiment_result_1['valid_acc'].apply(ast.literal_eval)
```

```python
plt.figure(figsize=(6, 5))

for index, row in experiment_result_1.iterrows():
    # Construct label with model details
    label = f"{row['model_type'].capitalize()} | {row['optimizer_type']} | Batch={row['batch_size']} | LR={row['learning_rate']:.0e}"
    # Plot each configuration
    plt.plot(row['train_loss'], label=label)

# Set plot title and labels
plt.title("Train Loss Changes - Experiment 1")
plt.xlabel("Epoch")
plt.ylabel("Train Loss")
# Place legend inside the plot area in the upper right corner
plt.legend(loc='upper right', fontsize=9)

plt.show()
```

# Experiment 3 - Coverage and stability

```python
experiment_result_3 = pd.read_csv('/content/drive/MyDrive/experiment_result_3_epchos.csv')
```

```python
experiment_result_3['train_loss'] = experiment_result_3['train_loss'].apply(ast.literal_eval)
experiment_result_3['valid_loss'] = experiment_result_3['valid_loss'].apply(ast.literal_eval)
experiment_result_3['train_acc'] = experiment_result_3['train_acc'].apply(ast.literal_eval)
experiment_result_3['valid_acc'] = experiment_result_3['valid_acc'].apply(ast.literal_eval)
```

```python
experiment_result_3.head()
```

```python
color_map = {
    'ResNet-18': cm.Blues,
    'MobileNet': cm.Greens,
    'AlexNet': cm.Reds
}
```

```python
plt.figure(figsize=(6, 5))

for index, row in experiment_result_3.iterrows():
    model_type = row['model_type']
    color_scheme = color_map[model_type]

    train_color = color_scheme(0.8)  #
    valid_color = color_scheme(0.6)  #

    plt.plot(row['train_loss'], color=train_color, label=f"{model_type} - Train")

    plt.plot(row['valid_loss'], color=valid_color, linestyle='--', label=f"{model_type} - Valid")

plt.title("Tuned Models Train and Validation Loss Changes - Experiment 3")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.legend(loc='upper right', fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(6, 5))

for index, row in experiment_result_3.iterrows():
    model_type = row['model_type']
    color_scheme = color_map[model_type]

    train_color = color_scheme(0.8)  #
    valid_color = color_scheme(0.6)  #

    plt.plot(row['train_acc'], color=train_color, label=f"{model_type} - Train")

    plt.plot(row['valid_acc'], color=valid_color, linestyle='--', label=f"{model_type} - Valid")

plt.title("Tuned Models Train and Validation Accuracy Changes - Experiment 3")
plt.xlabel("Epoch")
plt.ylabel("Acc%")

plt.legend(loc='lower right', fontsize=9)
plt.show()
```
